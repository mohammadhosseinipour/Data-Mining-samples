```python
def plot_series(dia_01):
    columns=dia_01.columns
    fig, axs = plt.subplots(4,3,figsize=(15,15))
#     fig.suptitle('lets double check')
    axs[0,0].hist(dia_01.iloc[:,0])
    axs[0,0].set_title(columns[0])
    axs[0,1].hist(dia_01.iloc[:,1])
    axs[0,1].set_title(columns[1])
    axs[0,2].hist(dia_01.iloc[:,2])
    axs[0,2].set_title(columns[2])
    axs[1,0].hist(dia_01.iloc[:,3])
    axs[1,0].set_title(columns[3])
    axs[1,1].hist(dia_01.iloc[:,4])
    axs[1,1].set_title(columns[4])
    axs[1,2].hist(dia_01.iloc[:,5])
    axs[1,2].set_title(columns[5])
    axs[2,0].hist(dia_01.iloc[:,6])
    axs[2,0].set_title(columns[6])
    axs[2,1].hist(dia_01.iloc[:,7])
    axs[2,1].set_title(columns[7])
    axs[2,2].hist(pd.to_numeric(dia_01.iloc[:,8]))
    axs[2,2].set_title(columns[8])
    axs[3,0].hist(pd.to_numeric(dia_01.iloc[:,8]))
    axs[3,0].set_title(columns[9])
    plt.show()
```

```python
plot_series(house)
```
