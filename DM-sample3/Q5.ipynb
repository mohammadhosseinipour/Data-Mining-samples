```python
thyroid_01=thyroid.copy()
```

```python
column1=thyroid.columns
thyroid_01["Serum_thyroxin"]=thyroid_01["Serum_thyroxin"].replace({"?":np.NaN})
thyroid_01["Serum_triiodothyronine"]=thyroid_01["Serum_triiodothyronine"].replace({"?":np.NaN})
thyroid_01["Basal_TSH"]=thyroid_01["Basal_TSH"].replace({"?":np.NaN})
thyroid_01["Serum_thyroxin"]=thyroid_01['Serum_thyroxin'].astype(str).astype(float)
thyroid_01["Serum_triiodothyronine"]=thyroid_01['Serum_triiodothyronine'].astype(str).astype(float)
thyroid_01["Basal_TSH"]=thyroid_01['Basal_TSH'].astype(str).astype(float)
for s in thyroid_01:
    thyroid_01[s]=thyroid_01[s].replace({np.NaN:pd.to_numeric(thyroid_01[s]).mean()})
```

```python
need=(0.25*(185)-30)/0.75
```

```python
to_resample=thyroid_01.loc[thyroid_01["Outcome"]==2]
```

```python
our_sample=to_resample.sample(n=int(need),replace=True)
```

```python
thyroid_reb=pd.concat([thyroid_01,our_sample])
thyroid_reb
```

```python
x=thyroid_reb.copy()
x=x.drop(["Outcome"],axis=1)
x.reset_index()
```

```python
y=thyroid_reb["Outcome"]
y=y.astype(int)
y.value_counts()
```

```python
t_train,t_test=train_test_split(thyroid_reb,test_size=0.2,random_state=7)
```

```python
t_train.shape
```

```python
t_test.shape
```

```python
t_x=t_test.copy()
t_x=t_x.drop(["Outcome"],axis=1)
t_x.reset_index()
```

```python
t_y=t_test["Outcome"]
t_y=t_y.astype(int)
t_y.value_counts()
```

```python
model=DecisionTreeClassifier(criterion="gini",max_depth=6).fit(x,y)
model
```

```python
tree.plot_tree(model)
plt.show()
```

```python
p_y=model.predict(t_x)
p_y=pd.Series(p_y)
p_y
```

```python
count=0
for i in range(0,len(p_y)):
    if p_y.iloc[i]==t_y.iloc[i]:
        count=count+1
        
acc=(count/len(p_y))*100
acc
```
